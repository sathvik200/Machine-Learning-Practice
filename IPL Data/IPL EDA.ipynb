<H2>Particulat player analysis

```python
batter_filter = deliveries_data['batter'].unique()
bowler_filter = deliveries_data['bowler'].unique()
def Particular_player_analysis(player_name):
    if(player_name in batter_filter):
        filt = (deliveries_data['batter'] == player_name)
        df_player = deliveries_data[filt]
        return df_player
    elif(player_name in bowler_filter):
        filt = (deliveries_data['bowler'] == player_name)
        df_player = deliveries_data[filt]
        return df_player
    else:
        return "NO player with such name"
```

```python
Particular_player_analysis('V Kohli')
```

<H2>Who are the leading run-scorers of all time?

```python
runs=deliveries_data.groupby(['batter'])['batsman_run'].sum().reset_index()
runs.columns=['Batsman','runs']
runs_batsmen=runs.sort_values(by='runs',ascending=False).head(10).reset_index()
```

```python
sns.barplot(x=runs_batsmen['Batsman'][:10], y=runs_batsmen['runs'][:10])
plt.title("Top 10 leading run-scorers", fontsize=24)
plt.xlabel("Player",fontsize=14)
plt.ylabel("Total Runs",fontsize=14)
plt.xticks(rotation=90)
plt.show()
```

<H2>Who has hit the most number of 4's?

```python
balls_played=balls_played.merge(runs,left_on='batter',right_on='Batsman',how='outer')
four=deliveries_data[deliveries_data['batsman_run']==4]
runs_4=four.groupby('batter')['batsman_run'].count().reset_index()
runs_4.columns=['Batsman','4s']
runs_4.sort_values(by='4s',ascending=False).head(10).reset_index().drop('index',axis=1)
```

<H2>Who has hit the most number of 6's?

```python
six=deliveries_data.groupby('batter')['batsman_run'].agg(lambda x: (x==6).sum()).reset_index()
six.columns=['Batsman','6s']
six.sort_values(by='6s',ascending=False).head(10).reset_index().drop('index',axis=1)
```

<H2>Who has the highest strike rate?

```python
player=pd.concat([runs,balls_played.iloc[:,1],runs_4.iloc[:,1],six.iloc[:,1]],axis=1)
player['strike_rate']=player['runs']/player['ballnumber']*100
player['4s'].fillna(0,inplace=True)
player.isnull().values.any()
```

```python
player.sort_values(by='strike_rate',ascending=False).head(10)
```

```python
strikerate=player[player.ballnumber > 100]
strikerate.sort_values(by='strike_rate',ascending=False).head(10)
reqsr=strikerate.drop(columns=['runs','ballnumber','4s','6s'],axis=1)
reqsr.sort_values(by='strike_rate',ascending=False).head(10)
```
